# Question 2

The following is a partial DNA sequence of the breast cancer biomarker gene BRAC2:

gggtgcgacg attcattgtt ttcggacaag tggataggca accactaccg gtggattgtc.

Using Python script:
- print out the sequence length and estimate its GC%
- extract the first 3 and the last 3 codons of the provided sequence
- check whether the extracted first codon is a start codon (atg) and the last codon is one of the stop codons (taa, tag, tga) or not.

```python
# Remove the inserted spaces between sequence bases and print the cleaned sequence (.replace) method
BRAC2_gene = "gggtgcgacg attcattgtt ttcggacaag tggataggca accactaccg gtggattgtc"

BRAC2_gene = BRAC2_gene.replace(" ", "")

print(BRAC2_gene)
```

```python
print("Sequence Length:", len(BRAC2_gene))
```

```python
gc = (BRAC2_gene.count('g') + BRAC2_gene.count('c')) / len(BRAC2_gene) * 100

if gc > 50:
  print(f'High GC Sequence found: {gc:.2f}%')
```

```python
print("First 3 Codons in the Sequence:", BRAC2_gene[:3])

print()

print("Last 3 Codons in the Sequence:", BRAC2_gene[-3::])
```

```python
# Define start and stop codons in variables as a string and list respectively
start_codon = "ATG"
stop_codons = ["TAA", "TAG", "TGA"]
```

```python
# Check and print if the first codon is a start codon (if..else)
if BRAC2_gene[:3].upper() == start_codon:
  print("First codon is a start codon")
else:
  print("Not a start codon!")
```

```python
# Check and print if the last codon is one of the stop codons (if..elif..else)
if BRAC2_gene[-3:].upper() == stop_codons[0]:
  print("Last codon is a stop codon!")
elif BRAC2_gene[-3:].upper() == stop_codons[1]:
  print("Last codon is a stop codon!")
elif BRAC2_gene[-3:].upper() == stop_codons[2]:
  print("Last codon is a stop codon!")
else:
  print("Last codon is NOT a stop codon!")
```
