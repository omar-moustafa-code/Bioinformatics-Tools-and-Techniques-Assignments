```python
from Bio import SeqIO, SeqUtils
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
```

```python
# 1. Initialize container variables
input_file = "/content/SARS-CoV-2_genome.gb"
output_file = "/content/sars_cov_2_protein.fasta"

protein_records = []  # list to store extracted CDS protein records
```

```python
# 2. Read the locally saved GenBank file
record = SeqIO.read(input_file, "genbank")

# 3. Calculate genome size and GC%
genome_length = len(record.seq)
gc_content = SeqUtils.gc_fraction(record.seq) * 100

# 4. Print genome information
sequence_id = record.id
organism = record.annotations.get("organism", "Unknown organism")

print(f"Total Genome Length: {genome_length} bp")
print(f"Overall GC Content: {gc_content:.2f}%")
print(f"GenBank ID: {sequence_id}")
print(f"Organism: {organism}")
print()
print("Extracting CDS and translate them into the corresponding protein sequences....")

# 5. Iterate through CDS features and extract protein data
for feature in record.features:
    if feature.type == "CDS":

        translation = feature.qualifiers.get("translation", [""])[0]
        protein_id = feature.qualifiers.get("protein_id", ["unknown_protein"])[0]
        product = feature.qualifiers.get("product", ["unknown_product"])[0]

        protein_record = SeqRecord(
            seq=Seq(translation),
            id=protein_id,
            description=product
        )

        # 6. Append to list
        protein_records.append(protein_record)

# 7. Write proteins to FASTA file
SeqIO.write(protein_records, output_file, "fasta")

print(f"Successfully wrote {len(protein_records)} CDS protein sequences to")
print(output_file)
```
