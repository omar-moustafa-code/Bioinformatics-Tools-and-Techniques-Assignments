## Part II: Examine the code and answer questions:

```python
from Bio import Entrez

Entrez.email = "omoustafa@aucegypt.edu"

# 1. Search the PROTEIN database
Search_term = "insulin[Protein Name] AND Danio rerio[Organism]"

search_handle = Entrez.esearch(db="protein", term=Search_term)
search_record = Entrez.read(search_handle)
search_handle.close()

# 2. Get the first ID
first_id = search_record["IdList"][0]

# 3. Fetch Summary using esummary(DocSum)
summary_handle = Entrez.esummary(db="protein", id=first_id)
summary_record = Entrez.read(summary_handle)
summary_handle.close()

# 4. Extract Data
# The summary_record is a list of dictionaries
info = summary_record[0]

# The protein accession is saved in вҖҳCaptionвҖҷ field of the summary_record
print(f"Protein Accession: {info['Caption']}")

# The protein length in amino acids is saved in вҖҳLengthвҖҷ field.
print(f"Length: {info['Length']} amino acids")
```

## Part III: Write the python code for the task (Bonus question)

```python
Entrez.email = "omoustafa@aucegypt.edu"

# PubMed search query
query = "Paabo S[Author] AND Neanderthal[Title/Abstract]"

print("Pubmed Query: Author field for Paabo and Title/Abstract for Neanderthal")

# Search PubMed
search_handle = Entrez.esearch(db="pubmed", term=query, retmax=100)
search_record = Entrez.read(search_handle)
search_handle.close()

# Get total count
count = search_record["Count"]
print("Total number of retrieved papers:", count)

# Extract IDs
search_ids = search_record["IdList"]

# Fetch abstracts
fetch_handle = Entrez.efetch(db="pubmed",
                             id=search_ids,
                             rettype="abstract",
                             retmode="text")

abstracts = fetch_handle.read()
fetch_handle.close()

print()

# Save abstracts to file
with open("All_papers_Paabo.txt", "w") as f:
    f.write(abstracts)

print("Records saved to All_papers_Paabo.txt")
```
