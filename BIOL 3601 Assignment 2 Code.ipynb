## Question 3

```python
rna_seq = "AUGCCGUUAGAC"
```

```python
# Print the total length of the RNA sequence using the len() function.
print("Sequence Length:", len(rna_seq))
```

```python
# Create a new variable dna_seq by converting the RNA back into DNA (replace all Uracil 'U' bases with Thymine 'T').

dna_seq = rna_seq.replace("U", "T")
print("cDNA Sequence:", dna_seq)
```

```python
# Adenine Count: Count the number of Adenine 'A' bases in the sequence using .count().
print("Adenine Count:", dna_seq.count('A'))
```

```python
# Report: Print the final DNA sequence and the Adenine count using an f-string.
print(f"Final DNA Sequence: {dna_seq} \nAdenine Count: {dna_seq.count('A')}")
```
