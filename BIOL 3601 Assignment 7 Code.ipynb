```python
from Bio.PDB import PDBParser
```

```python
parser = PDBParser()
structure = parser.get_structure("3AVE", "3AVE.pdb")

model = structure[0]

print("Number of chains:", len(list(model.get_chains())))
```

```python
chainA = model["A"]
chainB = model["B"]

resA = chainA[297]
resB = chainB[297]

# Use CA atom for distance
atomA = resA["CA"]
atomB = resB["CA"]

distance = atomA - atomB

print("Distance:", distance)
```
